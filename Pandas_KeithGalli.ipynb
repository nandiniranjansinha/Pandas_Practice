```python
# Find athletes born in cities that start with a vowel
vowel_cities = bios[bios["born_city"].str.contains(r'^[AEIOUaeiou]', na=False)]
```

```python
# Find athletes with names that contain exactly two vowels
two_vowels = bios[bios['name'].str.contains(r'^[^AEIOUaeiou]*[AEIOUaeiou][^AEIOUaeiou]*[AEIOUaeiou][^AEIOUaeiou]*$', na=False)]
```

```python
two_vowels
```

```python
# Find athletes with names that have repeated consecutive letters (e.g., "Aaron","Emmett")
repeated_letters = bios[bios['name'].str.contains(r'(.)\1', na=False)]
```

```python
repeated_letters
```

```python
# Finding athletes with names ending in 'son' or 'sen'
son_sen_names = bios[bios['name'].str.contains(r'son$|sen$', case = False, na = False)]
```

```python
son_sen_names
```

```python
# Finding athletes born in a year starting with '19'
born_19xx = bios[bios["born_date"].str.contains(r'^19', na = False)]
```

```python
# na = False
# handles no values in a row
```

```python
# Finding athletes with names that do not contain any vowels
no_vowels = bios[bios["name"].str.contains(r'^[^AIEOUaeiou]*$', na = False)]
```

```python
no_vowels
```

```python
# Finding athletes whose names contain a hyphen or an apostrophe
hyphen_apostrophe = bios[bios["name"].str.contains(r"[-']", na = False)]
```

```python
hyphen_apostrophe
```

```python
# Finding athletes with names that start and end with the same letter
start_end_same = bios[bios["name"].str.contains(r'^(.).*\1$', na = False)]
```

```python
start_end_same
```

```python
# Finding athletes with a born_city that has exactly 7 characters
city_seven_chars = bios[bios["born_city"].str.contains(r'^.{7}$',na = False)]
```

```python
city_seven_chars
```

```python
# Finding athletes with names containing three or more vowels
three_or_more_vowels = bios[bios["name"].str.contains(r'([AEIOUaeiou].*){3,}', case = False, na = False)]
```

```python
# lambda is quite helpful when dealing with specific categories
```

```python
def categorize_athlete(row):
  if row['height_cm'] < 175 and row['weight_kg'] < 70:
    return 'Lightweight'
  elif row['height_cm'] < 185 and row['weight_kg'] <= 80:
    return 'Middleweight'
  else:
    return 'Heavyweight'

bios['Category'] = bios.apply(categorize_athlete, axis = 1)
# axis = 1 means rows
```
